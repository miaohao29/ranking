```python
a=[]
with (open(Folder_Path +'\\'+ file_list[0],encoding='utf-8')) as f:
    z=re.sub(r'[a-z]*','',re.sub(r"(\W)(\d*)[a-z]*","",file_list[0]))#sub 应该是替换函数，用空替换掉所选择的字符
    y=re.sub('\d*',"",z)
    for li in f:
        a.append(li[:-1]+'\t'+ y +'\n')
for fi in file_list[1:]:
    b=[]
    with (open(Folder_Path +'\\'+ fi,encoding='utf-8')) as f:
        z=re.sub(r'[a-z]*','',re.sub(r"(\W)(\d*)[a-z]*","",fi))#sub 应该是替换函数，用空替换掉所选择的字符
        y=re.sub('\d*',"",z)
        print(y)
        for lin in f:
            b.append(lin[:-1]+'\t'+ y +'\n')
    print("befor",len(b))
    del b[0]
    print("after",len(b))
    a.extend(b)
    
```
