```python
#循环遍历列表中各个CSV文件名，并追加到合并后的文件
for i in range(1,len(file_list)):
    e=[]
    csv_reader=csv.reader(open(Folder_Path +'\\'+ file_list[i]))
    for row in csv_reader:
        e.append(row)#将读入的数据赋值给列表
    for index,li in enumerate(e[0]):
            if li.find("'")!=-1:
                f=e[0][index]
                g=f[f.find("'") + 1:]
                h=g[:g.find("'")]
                
    del e[:2]#删除两行表头
    del e[-2:]#删除最后两行            
                
    for row in e:
        row.append(h)#将期刊类别追加到每个文件的最后
        a.append(row)#将文件拼接到第一个文件后
```

```python
result=pd.DataFrame(np.array(a))#将pd.DataFrame(np.array(myarray))写入list
result.to_csv("C:/Users/rk/Desktop/2019jcr.csv",index=False,header=False,encoding='ANSI')
result
```
