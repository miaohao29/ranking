```python
#读入csv中的数据，（导入检索式）
a = []
csv_reader=csv.reader(open(Folder_Path +'\\'+ file_list[0]))#打开第一个文件
for row in csv_reader:
    a.append(row)#将读入的数据赋值给列表a
```

```python
del a[0]#删除一行表头
del a[-2:]#删除最后两行
```

```python
d=file_list[0][:file_list[0].find('.')]#从文件名获取学科类别
a[0].append('Categroy')
for item in a[1:]:
    item.append(d)
```

```python
h=file_list[2][:file_list[2].find('.')]
h
```

```python
#循环遍历列表中各个CSV文件名，并追加到合并后的文件
for i in range(1,len(file_list)):
    e=[]
    csv_reader=csv.reader(open(Folder_Path +'\\'+ file_list[i]))
    for row in csv_reader:
        e.append(row)#将读入的数据赋值给列表
    h=file_list[i][:file_list[i].find('.')]
    print(h)
    for row in e[2:-2]:
        row.append(h)#将期刊类别追加到每个文件的最后
        a.append(row)#将文件拼接到第一个文件后
print('every file has been saved')
```

```python
result=pd.DataFrame(np.array(a))#将pd.DataFrame(np.array(myarray))写入list
result.to_csv("C:/Users/rk/Desktop/category.csv",index=False,encoding='ANSI')
result
```
