```python
import csv
import copy
import numpy as np
import pandas as pd
```

```python
#读入csv中的数据，（导入检索式）
a = []
csv_reader=csv.reader(open(r'C:\Users\rk\Desktop\体院排名\puball.csv'))
for row in csv_reader:
    a.append(row)#将读入的数据赋值给列表a
```

```python
#使用跟这个进行拆分
null_list4=[]
for item in a[1:]:
    null_list5=item[1].split('[')#单条地址拆分
    for item_list5 in null_list5:
        temp=[]
        if item_list5!='':#地址中可能出现为零的情况，虽然网页上有
            temp.append(item[0])
            temp.append(item_list5)
            null_list4.append(temp)
```

```python
len(null_list4)
```

```python
null_list6=[]
for item in null_list4:
    temp=[]
    temp.append(item[0])
    p=item[1].find(']')
    temp.append(item[1][p+1:])
    template=copy.deepcopy(temp)
    null_list6.append(template)
```

```python
len(null_list6)
```

```python
null_list7=[]
for item in null_list6:
    null_list8=item[1].split(';')#单条地址拆分
    for item_list8 in null_list8:
        temp=[]
        if item_list8!=' ':#地址中可能出现为零的情况，虽然网页上有
            temp.append(item[0])
            temp.append(item_list8.strip())
            null_list7.append(temp)
```

```python
len(null_list7)
```

```python
result=pd.DataFrame(np.array(null_list7))#将pd.DataFrame(np.array(myarray))写入list
result.to_csv("C:/Users/rk/Desktop/spt7.csv",index=False,encoding='ANSI')
result
```
