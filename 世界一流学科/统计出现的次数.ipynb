```python
import re
import numpy as np, pandas as pd
import os
import csv
import time
```

```python
############首先对Mapp.csv原文件数据去重，删除数据重复的项目######################
lis_a = []#创建空白列表，存储文件内容
csv_reader=csv.reader(open(r'C:\Users\rk\Desktop\GRAS2019 Pub_CNCI_IC.csv'))
```

```python
for element in csv_reader:
    lis_a.append(element)#逐行读入数据，写入列表
```

```python
lis_b=[]
for element in lis_a:
    lis_b.append(element[0])
```

```python
t1=time.time()
list_c=[]
for element in lis_b:
    lis_d=[]
    lis_d.append(element)
    lis_d.append(lis_b.count(element))
    list_c.append(lis_d)
t2=time.time()
print(t1-t2)
```

```python
d=[]
for ite in list_c:
    if ite[1] != 1:
        d.append(ite)
```

```python
result=pd.DataFrame(np.array(d))#将pd.DataFrame(np.array(myarray))写入list
result.to_csv("C:/Users/rk/Desktop/"+'0426jk'+".csv",index=False,encoding='ANSI')
result
```
