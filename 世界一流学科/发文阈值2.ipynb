```python
####################运用此程序的前提是：先在原文件里做好分类汇总####################################
final_lis=[]
counter=0
while True:#为了是for循环中的起始值counter的改变，for进入循环后，起始值不再改变，
    for item_1 in c[counter:]:
        diction=[]
        lis_c=[]
        initial_s=0
        lis_c.append(item_1[1])#将第一个期刊加入列表，在此处加入是防止仅出现一次而忽略
        for item_2 in c[counter+1:]:
            if item_1[0]==item_2[0]:
                lis_c.append(item_2[1])#将相同的期刊加入列表
                initial_s+=1 #每找到一个相同的就把initial_s加一，     
        diction.append(item_1[0])
        diction.append(lis_c)
        final_lis.append(diction)
        break#为了使for循环中的起始值改变，循环一次后利用break强制退出for循环，利用while进入下一次循环
    counter=counter+initial_s+1
    if counter>=len(c):#为了使while循环跳出,当大于列表长度的时候跳出循环
        break
```

```python
def judge_mean(lis,mean,counter1,counter2,counter3):
    '''统计相应compare_numb的数量'''
    if float(lis[2])>=mean:
        counter1+=1
    else:
        counter2+=1
    if float(lis[2])>=mean*1.5:
        counter3+=1
    return counter1,counter2,counter3
```

```python
def judge(lis,compare_numb,counter,cnci):
    '''统计相应compare_numb的数量'''
    if int(lis[1])>=compare_numb:
        counter+=1
        cnci.append(lis[2])
    return counter,cnci
```

```python
def savedata(item,couter,cnci,subject):
    '''将统计的数量及CNCI最大值添加至当前列表，并判断统计数据是否为零'''
    subject.append(couter)
    subject.append(max(cnci))
    flag1=0
    for item_4 in item[1]:
        if max(cnci)!='0' and (max(cnci) in item_4):#判断当前发文阈值下，是否有学校数量counter及CNCI
            subject.append(item_4[0])
            flag1=1
            break
    if flag1== 0:
        subject.append('0')
    return subject,counter,cnci
```

```python
def mean_cnci(cnci):
    total=0
    for ite in cnci:
        total=float(ite)+total
        mean=total/len(cnci)
    return mean
```
