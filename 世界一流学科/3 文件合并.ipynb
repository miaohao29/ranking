```python
#读入csv中的数据，（导入检索式）
a = []
with open(Folder_Path +'\\'+ file_list[0],encoding='UTF8') as f:
    csv_reader=csv.reader(f)#打开第一个文件
    for row in csv_reader:
        a.append(row)#将读入的数据赋值给列表a
```
