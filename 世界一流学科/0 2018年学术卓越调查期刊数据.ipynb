```python
#_*_coding: utf-8_*_
import pandas as pd
import numpy as np
from selenium import webdriver #从selenium中调入webdriver
import time
```

```python
browser = webdriver.Chrome('C:/Users/rk/Desktop/chromedriver.exe') #锁定浏览器
browser.get("http://www.shanghairanking.com/shanghairanking_academic_excellence_survey_2017/index.jsp?param=ku6a5") #访问页面
```

```python
categorys=browser.find_elements_by_xpath('//*[@id="middle-wizard"]/div/div[2]/div/ul/ul/li/div/ins')
counter=0
all_category=[]
for i in range(57):
    categorys=browser.find_elements_by_xpath('//*[@id="middle-wizard"]/div/div[2]/div/ul/ul/li/div/ins')
    categorys2=browser.find_elements_by_xpath('//*[@id="middle-wizard"]/div/div[2]/div/ul/ul/li/label')
    categorys[i].click()
    category=categorys2[i].text
    while True:
        try:
            browser.find_element_by_xpath('//*[@id="surveybutton1"]').click()#点击前进
        except:
            time.sleep(1)
        else:
            break
    JCRS=browser.find_elements_by_xpath('//*[@id="journalcode"]/ul/li')#查询所有数据
    temp=[]
    for index,jcr in enumerate(JCRS):
        temp.append(jcr.text)
        if (index+1)%4==0:
            temp.append(category)
            all_category.append(temp)
            temp=[]
    print("waitting a moment...")
    time.sleep(1)
    counter+=1
    print(counter)
    while True:
        try:
            browser.find_element_by_xpath('//*[@id="bottom-wizard"]/button[1]').click()
        except:
            time.sleep(1)
        else:
            break
```
