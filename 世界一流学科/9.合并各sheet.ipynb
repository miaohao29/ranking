```python
import pandas as pd
```

```python
alldata=pd.ExcelFile(r"C:\Users\rk\Desktop\【计算表定稿】GRAS8-2020 排名计算表_20200618.xlsx")
```

```python
sheetnames=alldata.sheet_names
```

```python
result=alldata.parse('数学')
for item in alldata.sheet_names[4:]:
    sheet2=alldata.parse(item)
    result=pd.concat([result,sheet2],axis=0,join='inner')
    print(item)
```
