```python
#_*_coding: utf-8_*_
import pandas as pd
import numpy as np
from selenium import webdriver #从selenium中调入webdriver
```

```python
browser = webdriver.Chrome('C:/Users/rk/Desktop/chromedriver.exe') #锁定浏览器
browser.get("https://artsci.wustl.edu/faculty-staff/list#filter") #访问页面
```

```python
arry=browser.find_elements_by_xpath('//*[@id="content"]/section[4]/div[2]/a')#找出url
```

```python
url2=[]
for i in range(0,len(arry)):
    url2.append(arry[i].get_attribute('href'))#获取整个页面人员的URL信息
```

```python
url2[0]
```

```python
len(url2)
```

```python
counter=0
email_all=[]
```

```python
while True:
    for url3 in url2[counter:]:
        email_singlar=[]
        browser.get(url3)
        try:
            name=browser.find_element_by_xpath('//*[@id="content"]/article/div[1]/div[2]/div[2]/h1').text#姓名
        except:
            counter+=1
            break
        email_singlar.append(name)
        position=browser.find_element_by_xpath('//*[@id="content"]/article/div[1]/div[2]/div[2]/div[1]').text#职位
        email_singlar.append(position)
        email=browser.find_element_by_xpath('//*[@id="content"]/article/div[2]/div[1]/div[1]/ul/li[1]/a').text#邮箱
        email_singlar.append(email)
        email_singlar.append(url3)
        with open('C:/Users/rk/Desktop/em.txt','a+') as f:
            string='、'.join(email_singlar)
            f.write(string+'\n')
        print(email_singlar)
        email_all.append(email_singlar)
        counter+=1
        print(counter)
    if counter==len(url2):
        break
```

```python
counter+=1
```

```python
result=pd.DataFrame(np.array(email_all))#将pd.DataFrame(np.array(myarray))写入list
result.to_csv("C:/Users/rk/Desktop/email5.csv",index=False,encoding='ANSI')
result
```

```python
finalp=[]
for i in range(0,len(url2)):
    browser.get(url2[i])#访问每个学校
    name_l=browser.find_element_by_css_selector('h2.center').text#查找标题名称
    if len(name_l)==13:#如果标题有17个字符,说明还有分校
        arry2=browser.find_elements_by_xpath("//tbody//ul/li/a")
        url4=[]
        for j in range(0,len(arry2)):
            url4.append(arry2[j].get_attribute('href'))
        for k in range(0,len(url4)):
            browser.get(url4[k]) #访问分校
            name2=browser.find_element_by_css_selector('h2.center').text[5:-8]#获取分校名称
            arry3=browser.find_elements_by_xpath("//tbody//ul/li/a")
            url6=[]
            for j in range(0,len(arry3)):
                url6.append(arry3[j].get_attribute('href'))
            for b in range(0,len(url6)):
                browser.get(url6[b]) #访问省份
                rowlist = browser.find_elements_by_xpath('//tr')
                elemlist=[]
                for l in range(0,len(rowlist)):
                    elemlist.append(rowlist[l].find_elements_by_tag_name('td'))
                for m in range(1,len(elemlist)):#删除表头
                    p = []
                    for n in range(0,len(elemlist[m])):
                        p.append(elemlist[m][n].text)
                    p.append(name2)
                    finalp.append(p)
    else:
        arry4=browser.find_elements_by_xpath("//tbody//ul/li/a")#
        name3=browser.find_element_by_css_selector('h2.center').text[5:-8]
        url8=[]
        for f in range(0,len(arry4)):
            url8.append(arry4[f].get_attribute('href'))
        for g in range(0,len(url8)):
            browser.get(url8[g]) #访问页面
            rowlist = browser.find_elements_by_xpath('//tr')
            elemlist=[]
            for q in range(0,len(rowlist)):
                elemlist.append(rowlist[q].find_elements_by_tag_name('td'))
            for r in range(1,len(elemlist)):
                p = []
                for y in range(0,len(elemlist[r])):
                    p.append(elemlist[r][y].text)
                p.append(name3)
                finalp.append(p)
result=pd.DataFrame(np.array(finalp))#将pd.DataFrame(np.array(myarray))写入list
result.to_csv("C:/Users/rk/Desktop/2018年综合评价录取名单.csv",index=False,encoding='ANSI')
result
```
