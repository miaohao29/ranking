```python
import csv
```

```python
import numpy as np
import pandas as pd
```

```python
csv_reader=csv.reader(open(r'C:\Users\rk\Desktop\合并英文名.csv'))
```

```python
lis=[]
for element in csv_reader:
    lis.append(element)#逐行读入数据，写入列表
```

```python
lis
```

```python
len(lis[0][0])
```

```python
flag=0
temp=[]
fina=[]
start=1
```

```python
for item in lis[::-1]:
    if item[1]=='1':
        flag=1
        temp.append(item[0])
        #continue
    else:
        temp.append(item[0])
        flag=0
        start+=1
        #continue
    if flag==1:
        fina.append(temp)
        temp=[]
        flag=0
```

```python
fina
```

```python
for i in fina:
    while True:
        if len(i) != 7:
            i.append(',')
        else:
            break
```
