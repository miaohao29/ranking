```python
lis_b=[]
for item in lis_a[1:]:
    basic=copy.deepcopy(item[:4])
    tp1=item[4].split('！')
    #print(tp1)
    if len(tp1) == 1:#如果只有一个
        tp2=[]
        tp2.extend(basic)
        tp2.extend(['normal',tp1[0]])
        lis_b.append(copy.deepcopy(tp2))
        #print(tp2)
        continue
        
    if ('类' in tp1[0] or '班' in tp1[0]) and ('类' in tp1[1] or '班' in tp1[1])\
        or '中外合作' in tp1[1]:
        basic2=copy.deepcopy([tp1[0]+'（'+tp1[1]+'）'])
        mark=2
        #print(basic2)
    else:
        mark=0
        
    for tt in tp1[mark:]:
        tp3=[]
        #print(333)
        for item_sp in lis_c[1:]:
            if item_sp[0] == tt:
                if mark == 2:
                    tp3.extend(basic)
                    tp3.append('normal')
                    tp3.extend([basic2,tt])
                    lis_b.append(copy.deepcopy(tp3))
                    #print(tp3)
                else:
                    tp3.extend(basic)
                    tp3.extend(['normal',tt])
                    lis_b.append(copy.deepcopy(tp3))
                    #print(tp3)

            elif item_sp[0] in tt:
                if mark == 2:
                    tp3.extend(basic)
                    tp3.append('unnormal')
                    tp3.extend([basic2,tt])
                    lis_b.append(copy.deepcopy(tp3))
                   #print(tp3)
                else:
                    tp3.extend(basic)
                    tp3.extend(['unnormal',tt])
                    lis_b.append(copy.deepcopy(tp3))
                    #print(tp3)
```
