```python
#_*_coding: utf-8_*_
from selenium import webdriver #从selenium中调入webdriver
browser = webdriver.Chrome('C:/Users/rk/Desktop/chromedriver.exe') #锁定浏览器
browser.get("https://gaokao.chsi.com.cn/zsgs/zhpjlqmd--method-groupBySs,yxdm-10001,year-2018,zslx=3.dhtml") #访问页面
arry=browser.find_elements_by_tag_name('a')
url=[]
for i in range(0,len(arry)):
    url.append(arry[i].get_attribute('href'))#获取URL
    url2=[]
url2=url[4:(len(arry)-11)]#获取每个省份的URL
```

```python
df=[]
```

```python
for i in range(0,len(url2)):
    browser.get(url2[i])
    mystr2=browser.find_element_by_xpath('//tbody').text
    thelist=mystr2.split('\n')
    del thelist[0:3]
    myarray=[]
    for line in thelist:
        myelem = line.split(' ')#按照空格拆分列表
        myarray.append(myelem)
    import pandas as pd
    import numpy as np
    df.append(pd.DataFrame(np.array(myarray)))#讲pd.DataFrame(np.array(myarray))写入list
```

```python
df
```

```python
result=pd.concat(df)
result.to_csv("C:/Users/rk/Desktop/4.csv",index=False,encoding='ANSI')
```
