```python
from selenium import webdriver 
import time
import random
```

```python
from selenium.webdriver.common.keys import Keys
```

```python
import csv
```

```python
with open(r'C:\Users\rk\Desktop\univ.csv',encoding='utf-8') as f:
    li=csv.reader(f)
    a=list(li)
```

```python
a[:4]
```

```python
def write_record(item):
    '''保存每次检索结果，防止程序崩溃，浏览器崩溃，服务器奔溃'''
    with open(r'C:\Users\rk\Desktop\univ-1.txt','a+') as file:
            string='、'.join(item)
            file.write(string+"\n")
            #print(item)
```

```python
browser = webdriver.Chrome('C:/Users/rk/Desktop/chromedriver.exe') #锁定浏览器
```

```python
browser.get("https://gkcx.eol.cn/school/search")#访问网页
```

```python
counter=1
```

```python
for item in a[counter:]:
    browser.get("https://gkcx.eol.cn/school/search")#访问网页
    browser.find_element_by_id('headsearch').clear()
    browser.find_element_by_id('headsearch').send_keys(item[2])
    time.sleep(random.random()+1)
    browser.find_element_by_id('headsearch').send_keys(Keys.ENTER) 
    time.sleep(random.random()+2)
    
    univresult=browser.find_element_by_xpath('//*[@id="root"]/div/div/div/div/div/div/div[3]/div[2]/div[1]/span').text
    if univresult=='0':
        browser.get("https://gkcx.eol.cn/school/search")#访问网页
        continue #搜索学校，结果为0
    time.sleep(random.random()+2)
    trs=browser.find_elements_by_xpath('//*[@id="root"]/div/div/div/div/div/div/div[3]/div[2]/div[2]/div/ul/li')
    for tr in trs:
        univname=tr.find_element_by_xpath('div[1]/p[1]/span[1]').text
        print(univname)
        if univname==item[2]:
            tr.find_element_by_xpath('div[1]/p[1]/span[1]').click()
            break
    time.sleep(random.random()+2)
    try:
        browser.find_element_by_xpath('//*[@id="root"]/div/div/div/div/div/div/div[2]/div/div/ul/li[4]/span').click()#点击开设专业
    except:
        pass
    time.sleep(random.random()+3)

    
    trs=browser.find_elements_by_xpath('//*[@id="root"]/div/div/div/div/div/div/div[3]/div[1]/div[1]/div/div/div[3]/table/tbody/tr')
    if len(trs)==0:
        trs=browser.find_elements_by_xpath('//*[@id="root"]/div/div/div/div/div/div/div[3]/div[1]/div[1]/div/div/div[4]/table/tbody/tr')
        if len(trs)==0:
            trs=browser.find_elements_by_xpath('//*[@id="root"]/div/div/div/div/div/div/div[3]/div[1]/div[1]/div/div/div[2]/table/tbody/tr')
            if len(trs)==0:
                trs=browser.find_elements_by_xpath('//*[@id="root"]/div/div/div/div/div/div/div[3]/div/div[1]/div/div/div[2]/div[1]/div[2]/div[3]/table/tbody/tr')
    print(len(trs))
    for tr in trs[1:]:
        tds=tr.find_elements_by_xpath('td')
        cla=tds[0].text#专业类
        divs=tds[1].find_elements_by_xpath('div')
        for div in divs:
            tp=[]
            tp.extend(item) 
            tp.append(cla)
            special=div.text#专业
            tp.append(special)
            #print(tp)
            write_record(tp)
    counter+=1 
    print(counter)
    
```
