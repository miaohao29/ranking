```python
urlsubjects=driver.find_elements_by_xpath('/html/body/div[6]/div/section/div/div[1]/div/div/div/div/div/div/div/div/div/ul/li')
```

```python
subjects=[]
for urlsubject in urlsubjects:
    template=[]
    subjectname=urlsubject.find_element_by_xpath('div/span/a').text
    subjecturl=urlsubject.find_element_by_xpath('div/span/a').get_attribute('href')
    template.extend([subjectname,subjecturl])
    temp=copy.deepcopy(template)
    subjects.append(temp)
```

```python
def record(item):
    with open(r'C:\Users\rk\Desktop\theranking.txt','a+',encoding='utf-8') as f:
        strings='\t'.join(item)
        f.write(strings+'\n')
```

```python
def selectoption():
    '''点击显示全部'''
    try:
        driver.find_element_by_xpath('/html/body/div[6]/div/section/div/div[3]/div/div[1]/div[1]/div/div[3]/div[2]/label/select').click()
    except:
        driver.find_element_by_xpath('/html/body/div[6]/div/section/div/div[3]/div/div[1]/div[1]/div/div/div[2]/label/select').click()
    time.sleep(2+random.random())
    try:
        driver.find_element_by_xpath('/html/body/div[6]/div/section/div/div[3]/div/div[1]/div[1]/div/div[3]/div[2]/label/select/option[5]').click()
    except:
        driver.find_element_by_xpath('/html/body/div[6]/div/section/div/div[3]/div/div[1]/div[1]/div/div/div[2]/label/select/option[5]').click()
    time.sleep(1+random.random())
    driver.find_element_by_xpath('/html/body').click()#随便点一下，以确定上面点击的选项
    time.sleep(2+random.random())
```

```python
def getData(name,year):
    '''获取当前页面数据，为防止有空值，逐行获取数据'''
    #print(1)
    alluniversities=driver.find_elements_by_xpath('//table/tbody/tr')
    if len(alluniversities)==0:
        alluniversities=driver.find_elements_by_xpath('/html/body/div[6]/div/section/div/div[3]/div/div[1]/div[1]/div/div/table/tbody/tr')
    #print(2)    
    for university in alluniversities:
        #print(3)
        template=[]
        rank=university.find_element_by_xpath('td[1]').text
        informations=university.find_element_by_xpath('td[2]').text
        information=informations.split('\n')
        universityname=information[0]
        country=information[1]
        overall=university.find_element_by_xpath('td[3]').text
        citations=university.find_element_by_xpath('td[4]').text
        industryincome=university.find_element_by_xpath('td[5]').text
        internationaloutlook=university.find_element_by_xpath('td[6]').text
        research=university.find_element_by_xpath('td[7]').text
        teachering=university.find_element_by_xpath('td[8]').text
        template.extend([year,name,rank,universityname,country,overall,citations,industryincome,internationaloutlook,research,teachering])
        print(rank)
        #print(4)
        record(template)
        #print(5)
```

```python
def selectYear(n):
    '''选择年份'''
    driver.find_element_by_xpath('//*[@id="edit-jump"]').click()
    time.sleep(3+random.random())
    yearoptions=driver.find_elements_by_xpath('//*[@id="edit-jump"]/option')
    totalyear=len(yearoptions)#总共发布几年？
    year=yearoptions[n].text
    time.sleep(2+random.random())
    yearoptions[n].click()
    return year,totalyear
```

```python
for subject in subjects[start:]:
    driver.get(subject[1])
    year,totalyear=selectYear(0)
    for i in range(totalyear):
        #选择年份
        year,totalyear=selectYear(i)
        time.sleep(25+random.random()*10)
        #选择scores
        try:
            driver.find_element_by_xpath('/html/body/div[6]/div/section/div/div[3]/div/div[1]/div[1]/div/div[1]/ul/li[2]/label').click()#
        except:
            pass
        time.sleep(6+random.random()*5)
        #显示全部
        selectoption()
        #获取数据
        getData(subject[0],year)
    time.sleep(60+random.random()*30)
    start+=1
```
