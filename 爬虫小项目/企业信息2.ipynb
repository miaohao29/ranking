```python
#_*_coding: utf-8_*_
import pandas as pd
import numpy as np
from selenium import webdriver #从selenium中调入webdriver
import time
#from selenium.webdriver.common.keys import Keys
import random
```

```python
browser = webdriver.Chrome('C:/Users/rk/Desktop/chromedriver.exe') #锁定浏览器
```

```python
browser.get("https://www2.sgx.com/zh-hans/securities/corporate-information?page=1") #访问页面
```

```python
result=[]
```

```python
time.sleep(random.random()*2+4)
```

```python
random.random()
```

```python
page=1
```

```python
while True:
    print('page:',page)
    time.sleep(random.random()*2+4)
    allelement=browser.find_elements_by_xpath('//*[@id="page-container"]/template-base/div/div/sgx-widgets-wrapper/widget-filter-listing/widget-filter-listing-corporate-information/section[2]/div[2]/div/sgx-table/div/sgx-table-list/sgx-table-row')
    for element in allelement:
        temp=[]
        cop=element.find_element_by_tag_name('sgx-table-cell-link').text
        temp.append(cop)
        messages=element.find_elements_by_tag_name('sgx-table-cell-text')
        for mess in messages:
            temp.append(mess.text)
        result.append(temp)
        #print(temp)
    try:
        browser.find_element_by_xpath('//*[@id="page-container"]/template-base/div/div/sgx-widgets-wrapper/widget-filter-listing/widget-filter-listing-corporate-information/section[2]/div[2]/sgx-pagination/div/span[3]/span[1]/a').click()
        page+=1
    except:
        break
                
```

```python
final_result=[]
```

```python
for item in result:
    if item[0] != '':
        final_result.append(item)
        print(item)
        
```

```python
len(final_result)
```

```python
data=np.array(final_result)
```

```python
Datas=pd.DataFrame(data)
```
