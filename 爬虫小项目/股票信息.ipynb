```python
#_*_coding: utf-8_*_
import pandas as pd
import numpy as np
from selenium import webdriver #从selenium中调入webdriver
import time
from selenium.webdriver.common.keys import Keys
import random
```

```python
from selenium.webdriver.common.action_chains import ActionChains
```

```python
browser = webdriver.Chrome('C:/Users/rk/Desktop/chromedriver.exe') #锁定浏览器
```

```python
browser.get("https://www2.sgx.com/zh-hans/securities/securities-prices") #访问页面
```

```python
result=[]#初始化最终结果
dragdown_counter=0#滚动窗口的下拉次数
dragsum=0#滚动窗口的下拉总数
```

```python
#定位左右滚动条
Dragright = browser.find_element_by_xpath('//*[@id="page-container"]/template-base/div/div/sgx-widgets-wrapper/widget-securities-prices/section[1]/div[1]/sgx-table/sgx-scroll-shield/div[1]/div[2]/div')
#定位上下滚动条
Dragdown = browser.find_element_by_xpath('//*[@id="page-container"]/template-base/div/div/sgx-widgets-wrapper/widget-securities-prices/section[1]/div[1]/sgx-table/sgx-scroll-shield/div[2]/div[2]/div')
name_all=[]#存放公司名称，为了生成唯一的列表，方便判断是否该存储数据
while True:
    print("process0")
    counter=0
    date_element=browser.find_elements_by_xpath('//sgx-table-list/sgx-table-row')#查找每一行
    for date_ele in date_element:
        null_1=[]
        #time.sleep(1)
        ActionChains(browser).drag_and_drop_by_offset(Dragright, -500, 0).perform()
        name=date_ele.find_element_by_tag_name('sgx-table-cell-securities').text
        name_all.append('intial_name')#初始化公司列表，防止后面无法判断
        null_1.append(name)
        code=date_ele.find_element_by_tag_name('sgx-table-cell-text').text
        null_1.append(code)
        numbers=date_ele.find_elements_by_tag_name('sgx-table-cell-number')
        #for nu in numbers[:-6]:
        #print("process1")
        for nu in numbers:
            null_1.append(nu.text)
'''下面被注销的几行代码：由于每行都得暂停，所以，耗费时间特别多，可以分两步抓取数据，或者调整网页显示'''
#         ActionChains(browser).drag_and_drop_by_offset(Dragright, 500, 0).perform()#向右滚动条
#         time.sleep(1)
#         numbers1=date_ele.find_elements_by_tag_name('sgx-table-cell-number')#重新查找每行数据
#         for n in numbers1[-6:-3]:#记录后几列数据
#             null_1.append(n.text)
        
        name_final=np.unique(name_all)#生成唯一性的公司列表
        print(len(name_final))#列表中存在公司个数
        #print("process2")
        if (name not in name_final) and (name !=''):#判断该行是否存在数据，若存在，则该公司数据是否已存储，若存在数据且该公司数据未3存储，则记录该公司数据
            result.append(null_1)
            print(null_1) 
            counter+=1
        name_all.append(name)#将公司名添加到公司列表，只能在上一步之后
        counter+=1
    if dragsum>=10:#若整体像素超过该值，则跳出循环
        break
    ActionChains(browser).drag_and_drop_by_offset(Dragdown, 0, 6).perform()#下拉滚动条
    dragdown_counter+=1
    dragsum=dragdown_counter*6#总共的滚动距离
    time.sleep(3)
```

```python
result_final=pd.DataFrame(np.array(result))
result_final.to_csv("C:/Users/rk/Desktop/result_final4.csv",index=False,encoding='ANSI')
result_final
```
