```python
#_*_coding: utf-8_*_
import pandas as pd
import numpy as np
from selenium import webdriver #从selenium中调入webdriver
import time
from selenium.webdriver.common.keys import Keys
import random
from selenium.webdriver.common.action_chains import ActionChains
```

```python
browser = webdriver.Chrome('C:/Users/rk/Desktop/chromedriver.exe') #锁定浏览器
```

```python
browser.get("http://global.krx.co.kr/board/GLB0501050100/bbs") #访问页面
```

```python
result=[]
page=1
```

```python
while True:
    data_element=browser.find_elements_by_xpath('//tbody/tr')
    for data_ele in data_element:
        tr_null=[]
        data_td=data_ele.find_elements_by_tag_name('td')
        flag=data_td[0].text
        for data_tr in data_td:
            tr_null.append(data_tr.text)
        print(tr_null)
        result.append(tr_null)
    page+=1
    print(page)
    try:
        temp=browser.find_elements_by_xpath('//nav/ul/li[1]/ul/li')
        #print(6)
    except:
        temp=browser.find_elements_by_xpath('//nav/ul/li[3]/ul/li')
        #print(5)
    for tep in temp:
        #print(7)
        if tep.text==str(page):
            #print(1)
            tep.click()
            break
            #print('heeee')
    time.sleep(5)
    time.sleep(random.random()*5)
    if page==177:
        break
```
