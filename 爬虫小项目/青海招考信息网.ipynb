```python
#_*_coding: utf-8_*_
import pandas as pd
import numpy as np
from selenium import webdriver #从selenium中调入webdriver
import time
```

```python
browser = webdriver.Chrome('C:/Users/rk/Desktop/chromedriver.exe') #锁定浏览器
```

```python
browser.get("http://www.qhzk.com/YXLQShow.action") #访问页面
```

```python
#账号：13787240607
```

```python
result=[]
```

```python
currentpage=browser.find_elements_by_xpath('//*[@id="GridView1"]/tbody/tr')
```

```python
def get_datas(currentpage):
    #global result
    for current in currentpage:
        trdatas=current.find_elements_by_xpath('td')
        template=[]
        for trdata in trdatas:
            template.append(trdata.text)
        #return(template)
        result.append(template)
        print(template)
```

```python
def get_next_page():
    while True:
        global flag
        try:
            browser.find_element_by_link_text('下一页').click()
            time.sleep(3)
            break
        except:
            time.sleep(1)
            page=browser.find_element_by_xpath('//*[@id="AspNetPager1"]/div[1]/font').text
            if page == '128':
                flag=1
                break
```

```python
while True:
    flag=0
    currentpage=browser.find_elements_by_xpath('//*[@id="GridView1"]/tbody/tr')
    get_datas(currentpage)
    get_next_page()
    if flag==1:
        break
```
