```python
import pandas as pd
```

```python
import csv
```

```python
import copy
```

```python
c=[]
with open(r'C:\Users\rk\Desktop\TCC_2.csv') as f:
    data_2=csv.reader(f)
    for item in data_2:
        c.append(item)
```

```python
c[1:4]
```

```python
final_result=[]
```

```python
for item in c:
    print(item)
    temp_1=[]
    temp_1=item[:2]
    print(temp_1)
    splitings=item[2].split(',')#按照分号分列
    print(splitings)
    for spliting in splitings:
        spliting=spliting.strip()
        #print(spliting)
        temp_2=[]
        temp_2.extend(temp_1)
        temp_2.append(spliting)
        temp_3=copy.deepcopy(temp_2)
        final_result.append(temp_3)
```

```python
final_result
```

```python
def writeRecord(item):
    strings='、'.join(item)
    with open(r'C:\Users\rk\Desktop\TCC_2.txt','a+',encoding='utf-8') as f:
        f.write(strings+'\n')
```

```python
for item in final_result:
    writeRecord(item)
```
