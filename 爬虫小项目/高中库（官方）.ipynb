```python
def get_data(name):
    rows=driver.find_elements_by_xpath('/html/body/div/div[2]/div[1]/table[2]/tbody/tr')
    for row in rows:
        colums=row.find_elements_by_xpath('td')
        for col in colums:
            schoolname=col.text
            try:
                schoolurl=col.find_element_by_xpath('a').get_attribute('href')
            except:
                schoolurl="NA"
            school.append([name[0],schoolname,schoolurl])
            record([name[0],schoolname,schoolurl])
            print(school[-1])
```
