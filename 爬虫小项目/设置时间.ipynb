```python
#_*_coding: utf-8_*_
import pandas as pd
import numpy as np
import urllib
from bs4 import BeautifulSoup
from urllib.parse import urljoin

myurl= 'https://gaokao.chsi.com.cn/zsgs/zzlqmd--method-groupBySs,yxdm-10002,year-2018,zslx=1.dhtml'
fp = urllib.request.urlopen(myurl)#发送请求并且保存为文件
mybytes=fp.read() #() #读取该文件，内容为原始编码
myhtml= mybytes.decode("utf8")#解码
fp.close() #关闭文件
soup = BeautifulSoup(myhtml, 'html.parser')  #用bs4分析html脚本
```

```python
url_pre=soup('a')
n=len(url_pre)
```

```python
url=[]
for i in range(4,(n-11)):
    url.append(urljoin(myurl,url_pre[i].get('href')))#利用url函数拼接URL
```

```python
len(url)
```

```python
import time
```

```python
time.time()
```

```python
m=len(url)
final=[]
for i in range(0,m):
    time_0 = time.time()
    fp = urllib.request.urlopen(url[i])#发送请求并且保存为文件
    mybytes=fp.read() #() #读取该文件，内容为原始编码
    myhtml= mybytes.decode("utf8")#解码
    fp.close() #关闭文件
    soup = BeautifulSoup(myhtml, 'html.parser')  #用bs4分析html脚本
    time_1 = time.time()
    print('urlrequest:',time_1-time_0)
    td_list = soup.find_all('td') #找到html中的所有tr元素
    
    final_text=[]
    for td in td_list:
        final_text.append(td.get_text())
    
    del final_text[0:14]
    
    time_2 = time.time()
    print('textappend_1',time_2-time_1)
    p=[]
    for i in range(0,len(final_text)):
        p.append(final_text[i])
        if final_text[i]=='\n':
            final.append(p)
            p=[]
    time_3 = time.time()
    print('textappend_2',time_3-time_2)
    
    print('=====overall',time_3-time_0)
    
    
    
    
```

```python
result=pd.DataFrame(np.array(final))#将pd.DataFrame(np.array(myarray))写入list
```
