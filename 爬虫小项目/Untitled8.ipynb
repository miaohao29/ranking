```python
browser = webdriver.Chrome('C:/Users/rk/Desktop/chromedriver.exe') #锁定浏览器
browser.get("http://www.iwencai.com/stockpick/search?typed=1&preParams=&ts=1&f=1&qs=result_rewrite&selfsectsn=&querytype=hkstock&searchfilter=&tid=stockpick&w=%E4%B8%9C%E6%96%B9%E8%AF%81%E5%88%B8+")#访问网页
```

```python
count=0
```

```python
for i in range(x)[count:]:
    browser.find_element_by_xpath('//*[@id="auto"]').clear()
    browser.find_element_by_xpath('//*[@id="auto"]').send_keys(data.iloc[i,2])#输入公司名称
    browser.find_element_by_xpath('//*[@id="qs-enter"]').click()#点击搜索
    time.sleep(random.random()*2+1)
    try:
        browser.find_element_by_xpath('//*[@id="tableWrap"]/div[2]/div/div[2]/div/table/tbody/tr/td[4]/div/a').click()#点击股票简称
    except:
        data.iloc[i,20]= '无公司信息'
        continue
    
    all_windows = browser.window_handles
    browser.switch_to.window(all_windows[-1])
    
    while True:
        try:
            url = browser.find_element_by_xpath('/html/body/div[9]/div/ul[6]/li/a').get_attribute('href')#获得当前页面的高管信息
            break
        except:
            time.sleep(0.5)
            print('waiting...')
            pass
        
    data.iloc[i,19]= url
    time.sleep(random.random()+1)
    browser.get(url)#点击高管
    time.sleep(random.random()*2+1)
    
    tables = browser.find_elements_by_xpath('//*[@id="introduce"]/div[2]/table')#所有高管信息
    for table in tables:
        flag=0
        name = table.find_element_by_xpath('thead/tr[1]/td[1]/h3').text #高管姓名
       # print(name)
        if data.iloc[i,4] == name:
            flag=1
            data.iloc[i,20]= '找到'
            break
    if flag==0:
        data.iloc[i,20]= '未找到'
    time.sleep(random.random()+0.3)
    browser.close()
    all_windows = browser.window_handles
    browser.switch_to.window(all_windows[-1])
    time.sleep(random.random()*2+1)
    #position = tables[1].find_element_by_xpath('thead/tr[1]/td[2]').text#高管职位
    print(data.iloc[i,-1])
    data.to_csv(r'C:\Users\rk\Desktop\港交所结果追加模式.csv',mode="a",index=False,encoding="GB18030")
    count+=1
    print(count)
```
